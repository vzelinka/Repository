> згенеруємо декілька масивів різними способами та виконаємо базові математичні операції.

```python
l = (3, 3)

O = np.ones(l)
R = np.random.random(l)
Z = np.zeros(l)
C = np.arange(l[0])

print(f"""
Віднімання:\n {C - (O + 10)}
Множення:\n {O * R}
Ділення:\n {R / C}
Степінь:\n {C ** Z}
Поріняння:\n {R > 0.5}
Ділення на 0:\n {O / Z}
""")
E = np.eye(l[0])
print(f"""
Створена одинична матриця E:
{E}
Порівняння значень на наявність 1:
{E == 1}
""")
ones_matrix = np.ones(l)
result_matrix = ones_matrix - E
print(f"""
Матриця з одиниць:
{ones_matrix}
Матриця з одиничною діагоналлю:
{E}
Результат віднімання одиничної діагональної матриці від матриці з одиниць:
{result_matrix}
""")
```

# Множення матриць

```python
A = np.array([ [1, 2, 3], [2, 2, 2], [3, 3, 3] ])
B = np.array([ [3, 2, 1], [1, 2, 3], [-1, -2, -3] ])
print(f"Маємо два масиви:\n {A} \n {B}")

R = A * B
print(f"Тут виконується поелементне множення:\n {R}")

R = np.dot(A, B)
print(f"Тут, масиви перемножуються як матриці:\n {R}")

MA = np.asmatrix(A)
MB = np.asmatrix(B)

R = MA * MB
print(f"Тут масиви перетворені на матриці і далі перемножуються:\n {R}")
```

# Вирівнювання розмірностей

```python
A = np.array([ [11, 12, 13], [21, 22, 23], [31, 32, 33] ])
B = np.array([1, 2, 3])
C = B[:, np.newaxis]

print(f"Вектор В є рядком {B.shape}: {B} тому додавання буде по рядках:\n {A + B}")

print(f"""Вектор C є стовпцем {C.shape}:
{C}
тому додавання буде по стовпцям:
{A + C}""")

print(f"У випадку додавання рядка до стовпця буде матриця в якої доповнюються рядки і стовпці:\n {B + C}")
```

# Конкатенація, повторення та зміна форми (shape)

```python
import numpy as np

D = np.ones(3)
Z = np.zeros((3, 1))
B = np.array([[3, 2, 1], [1, 2, 3], [-1, -2, -3]])

D_reshaped = D[:, np.newaxis] 

print(f"""
При конкатенації рядків маємо: {np.concatenate((D_reshaped, D_reshaped), axis=0)}
При конкатенації стовпців маємо:\n{np.concatenate((Z, Z), axis=1)}
Масив D {D} має розмірність {D.shape} тому може виконати операцію конкатенації з В {B} розмірністю {B.shape}:
{np.concatenate((B, D_reshaped), axis=1)}  # Concatenating B and reshaped D along columns
Для конкатенації з С
{B}
де С має розмірність = {B.shape} потрібно змінити розмірність D = {D.shape}
Це можна ось так:
{D[:, np.newaxis].shape}
або
{D.reshape((3, 1)).shape}
в результаті ми можем виконати конкатенацію:
{np.concatenate((B, D.reshape((3, 1))), axis=1)}  # Concatenate B and reshaped D along columns
При повторенні D {D} 2 рази для рядкі та 3 для стовпців будем мати:
{np.tile(D, (2, 3))}
Як бачимо конкатенація та повторення є дещо схожими у деяких випадках:
- конкатенація: {np.concatenate((D, D))}
- повторення: {np.tile(D, 2)}
Також можна перетворити двовимірний масив A розмірністю {B.shape} на рядок: {B.flatten()} з розмірністю {B.flatten().shape}
""")

array_2d = np.arange(6).reshape(2, 3)
print(f"Створений масив за допомогою arange:\n{array_2d}")

concatenated_rows = np.concatenate((array_2d, array_2d), axis=0)
concatenated_columns = np.concatenate((array_2d, array_2d), axis=1)
print(f"Конкатенація по рядках:\n{concatenated_rows}")
print(f"Конкатенація по стовпцях:\n{concatenated_columns}")

tiled_array = np.tile(array_2d, (2, 2))
print(f"Повторення масиву:\n{tiled_array}")

array_different_shape = np.arange(9).reshape(3, 3)
print(f"Масив з різною кількістю рядків та стовпців:\n{array_different_shape}")

reshaped_array = array_2d.reshape(3, 2)
concatenated_with_reshape = np.concatenate((reshaped_array, array_different_shape), axis=1)
print(f"Конкатенація з перетворенням розмірності:\n{concatenated_with_reshape}")

array_1x12 = np.arange(12).reshape(1, 12)
array_2x6 = np.arange(12).reshape(2, 6)
array_3x4 = np.arange(12).reshape(3, 4)
array_4x3 = np.arange(12).reshape(4, 3)
array_6x2 = np.arange(12).reshape(6, 2)
array_12x1 = np.arange(12).reshape(12, 1)

print(f"""
Можливі варіанти reshape для масиву з 12 елементів:
1x12:\n{array_1x12}
2x6:\n{array_2x6}
3x4:\n{array_3x4}
4x3:\n{array_4x3}
6x2:\n{array_6x2}
12x1:\n{array_12x1}
""")
```

# Трансформації масивів

```python
O = np.arange(6).reshape(3,2)
O_r = O.reshape((2,3))
O_f = O.flatten()
O_t = O.transpose()
O_fl = np.flip(O)
O_fa = np.flip(O, axis=1)

print(f"""{O}
Змінити розмірність shape {O.shape} на {O_r.shape} при цьому розмір size {O.size} = {O_r.size} та ndim {np.ndim(O)} = {np.ndim(O_r)}:
{O_r}
Також можна перетворити {np.ndim(O)}-х вимірний масив розмірністю {O.shape} на рядок:
{O_f} 
з розмірністю {O_f.shape} при цьому size = {O_f.size} та ndim стане {np.ndim(O_f)}-х вимірним 
Або транспонувати масив щоб він став {O_t.shape} але зверніть увагу на елементи:
{O_t}
Можна обернути масив (обертаємо по стовпцях і рядках одночасно): 
{O_fl}
Або обернути лише елементи в кожному рядку (або стовпці):
{O_fa}
""")

vector_12 = np.arange(12)
print(f"Вектор з 12 елементів:\n{vector_12}")
print(f"shape: {vector_12.shape}, size: {vector_12.size}, ndim: {np.ndim(vector_12)}")

matrix_3x4 = vector_12.reshape(3, 4)
print(f"Матриця 3x4:\n{matrix_3x4}")
print(f"shape: {matrix_3x4.shape}, size: {matrix_3x4.size}, ndim: {np.ndim(matrix_3x4)}")

transposed_matrix = matrix_3x4.T
print(f"Транспонована матриця:\n{transposed_matrix}")
print(f"shape: {transposed_matrix.shape}, size: {transposed_matrix.size}, ndim: {np.ndim(transposed_matrix)}")

flipped_matrix = np.flip(transposed_matrix)
print(f"Обернена матриця:\n{flipped_matrix}")
print(f"shape: {flipped_matrix.shape}, size: {flipped_matrix.size}, ndim: {np.ndim(flipped_matrix)}")

flattened_array = flipped_matrix.flatten()
print(f"Перетворений назад вектор:\n{flattened_array}")
print(f"shape: {flattened_array.shape}, size: {flattened_array.size}, ndim: {np.ndim(flattened_array)}")
```
